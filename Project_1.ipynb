# Task 8 💀
Создайте переменную `has_kids`, которая принимает значение `True`, если клиент при бронировании указал хотя бы одного ребенка `(total_kids)`, в противном случае – `False`. 

Посчитайте отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах `(churn rate)`. Укажите, среди какой группы показатель выше

```python
data['has_kids']=np.where(data['total_kids']>0.0,True,False)
```

```python
data.head()
```

```python
data['has_kids'].value_counts()
```

```python
all_clients=data.shape[0]
canceled=data[data['is_canceled']==1].shape[0]
```

```python
churn_rate=(canceled/all_clients)*100
churn_rate
```

```python
# Clients with kids
t=data[data['has_kids']==1]
# Clients without kids
f=data[data['has_kids']==0]
```

`churn rate` c детьми

```python
cancel_h_kids=t[t['is_canceled']==1].shape[0]
all_h_kids=t.shape[0]
```

```python
churn_rate_with_kids=round((cancel_h_kids/all_h_kids)*100, 2)
churn_rate_with_kids
```

`churn rate` без детей

```python
cancel_no_kids=f[f['is_canceled']==1].shape[0]
all_no_kids=f.shape[0]
```

```python
churn_rate_without_kids=round((cancel_no_kids/all_no_kids)*100, 2)
churn_rate_without_kids
```
